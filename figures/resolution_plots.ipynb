```python
import matplotlib.pyplot as plt
import numpy as np
import torch
class Config: pass
```

```python
seq_length = [1, 2, 3]
pixels_acc = [23.104, 55.794, 71.13]
walloc_acc = [50.266, 64.336, np.nan]
fig = plt.figure(figsize=(3, 2), dpi=150)
plt.plot(seq_length, pixels_acc, marker='o', label='Pixels', color='black')
ax = plt.gca()
ax.plot(seq_length, walloc_acc, marker=(4, 1, 45), label='WaLLoC', color='blue', markersize=10)
ax.set_title('Image classification')
ax.set_xlabel('Resolution equivalent')
ax.set_ylabel('Accuracy (%)')
ax.grid(True)
ax.set_ylim([20,75])
ax.set_yticks([25,50,75])
ax.set_xticks(seq_length)
ax.set_xticklabels(['$64^2$ px', '$128^2$ px', '$256^2$ px'])
ax.legend(loc='lower center')
fig.savefig('classification.svg')
```
